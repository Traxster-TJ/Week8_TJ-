```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Name the columns from the auto_names file
```

```python
columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 
           'acceleration', 'model_year', 'origin', 'car_name']
```

```python
# Use the sep=r'\s+' command because delim_Whitespace was deprecated
```

```python
df = pd.read_csv(r"C:\Users\WIN\Downloads\Data\auto-mpg.data", sep=r'\s+', names=columns)
```

```python
#Replace the NaN values with pandas missing value syntax 
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'].replace('?', pd.NA))
```

```python
# Convert the values on the columns to match the respective countries
```

```python
Auto_Countries = {1: 'USA', 2: 'Europe', 3: 'Asia'}
df['origin'] = df['origin'].map(Auto_Countries)
```

```python
print("Automobiles:")
display(df.head())
```

```python
#Create a barchart for Cylinder distribution
```

```python
plt.figure(figsize=(4, 4))
df['cylinders'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Cylinders in Cars')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
#Create a scatterplot for Horsepower X Weight
```

```python
plt.figure(figsize=(4, 4))
sns.scatterplot(data=df, x='horsepower', y='weight')
plt.title('Horsepower X Weight')
plt.show()
```

```python
# Does the Data hold up to what we know in real life about the avg MPG per country?
```

```python
plt.figure(figsize=(4, 4))
sns.boxplot(data=df, x='origin', y='mpg')
plt.title('MPG Distribution by Region')
plt.show()
```
